```python
import cirq
import numpy as np
from cirq.contrib.svg import SVGCircuit as draw
def bitstring(bits):
    return ''.join(str(int(b)) for b in bits)
def binary_labels(num_qubits):
    return [bin(x)[2:].zfill(num_qubits) for x in range(2 ** num_qubits)]
```

```python
q=cirq.LineQubit.range(3)
c=cirq.Circuit()
```

```python
c.append(cirq.H(q[1]))
c.append(cirq.CNOT(q[1],q[2]))
```

```python
s = cirq.Simulator()
result = s.simulate(c, qubit_order=[q[0], q[1],q[2]])
result.dirac_notation()
```

```python
c.append(cirq.CNOT(q[0],q[1]))
draw(c)
```

```python
c.append(cirq.H(q[0]))
draw(c)
```

```python
c.append(cirq.measure(q[0],q[1]))
draw(c)
```

```python
c.append(cirq.CNOT(q[1],q[2]))
draw(c)
```

```python
c.append(cirq.CZ(q[0],q[2]))
draw(c)
```

```python
c.append(cirq.measure(q[2],key='BOB'))
draw(c)
```

```python
s = cirq.Simulator()
sample=s.run(c,repetitions=10000)
```

```python
cirq.plot_state_histogram(sample,tick_label=binary_labels(3))
```

```python
result=sample.histogram(key='BOB',fold_func=bitstring)
```

```python
cirq.plot_state_histogram(result)
```
